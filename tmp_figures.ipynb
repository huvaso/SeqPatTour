```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

# Basic Librariess
import numpy as np
import pandas as pd
import matplotlib as plt
from matplotlib import rc
import matplotlib.pyplot as plt

# Other libraries
import warnings
warnings.filterwarnings('ignore')
```

```python
# Set the text.usetex parameter to True to enable LaTeX rendering
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"]
})

plt.rcParams.update({'font.size': 13})
```

```python
## Construction of dataset for visualization purposes

data_path = []
t = ['A-star', 'BFS', 'Dijkstra', 'Dijkstra_nx', 'Shortest_nx', 'Dijkstra_ni']
time = [0.7243831250000312,4.547183832999963,0.005659917000002679,0.029057125000008455,0.0041207499998563435,0.2209250000000793]
size = [2172, 2310, 2310, 2311, 2311, 2310]
data_path.append(t)
data_path.append(time)
data_path.append(size)
```

```python
## Ploting the execution time par path extraction technique 

names = data_path[0][:3]
time = data_path[1][:3]
npaths = data_path[2][:3]


fig, ax = plt.subplots()
bar_container = ax.bar(names, time)
ax.set_ylabel('Execution time in seconds')
ax.set_xlabel('Path extraction technique')
ax.bar_label(bar_container, fmt='{:,.3f}')
#plt.grid(True)
#ax.grid(axis='y')
ax.tick_params(labelsize=10)
xmin, xmax = ax.get_xlim()
ax.set(xlim=(xmin-0.25, xmax+0.25), axisbelow=True)
ax.grid(axis='y')
# Save the plot as a PDF file
plt.savefig('bars_time_paths.pdf', bbox_inches='tight')
plt.show()
```

```python
## Ploting the number of patterns par path extraction technique 

names = data_path[0][:3]
time = data_path[1][:3]
npaths = data_path[2][:3]


fig, ax = plt.subplots()
bar_container = ax.bar(names, npaths)
ax.set_ylabel('Number of paths')
ax.set_xlabel('Path extraction technique')
ax.bar_label(bar_container, fmt='{: .0f}')
#plt.grid(True)
#ax.grid(axis='y')
ax.tick_params(labelsize=10)
xmin, xmax = ax.get_xlim()
ax.set(xlim=(xmin-0.25, xmax+0.25), axisbelow=True)
ax.grid(axis='y')
# Save the plot as a PDF file
plt.savefig('bars_number_paths.pdf', bbox_inches='tight')
plt.show()
```
